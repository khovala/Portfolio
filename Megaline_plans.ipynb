```python
# –ü–æ—Å—á–∏—Ç–∞–µ–º –≤—ã—Ä—É—á–∫—É —Å –∫–∞–∂–¥–æ–≥–æ –∞–±–æ–Ω–µ–Ω—Ç–∞ –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü.

def extra_cost (row):
    cost_sum=0
    cost1=row['duration']
    cost2=row['gb_used']
    cost3=row['messages']
    cost4=row['minutes_included']
    cost5=row['mb_per_month_included']/1024
    cost6=row['messages_included']
    cost7=row['rub_per_minute']
    cost8=row['rub_per_gb']
    cost9=row['rub_per_message']
    if cost1 > cost4:
        cost_sum += ((cost1-cost4)*cost7)
    if cost2 > cost5:
        cost_sum += ((cost2-cost5)*cost8)
    if cost3 > cost6:
        cost_sum += ((cost3-cost6)*cost9)
    return cost_sum
data_total['cost_per_month']=data_total.apply(extra_cost, axis=1) + data_total['rub_monthly_fee']
display(data_total.head(10))
display(data_total.tail(10))

        
```

## –®–∞–≥ 5. –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑

‚Ä¢	—Å—Ä–µ–¥–Ω—è—è –≤—ã—Ä—É—á–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ –ú–æ—Å–∫–≤—ã –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –≤—ã—Ä—É—á–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ –¥—Ä—É–≥–∏—Ö —Ä–µ–≥–∏–æ–Ω–æ–≤.

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –°—Ä–µ–¥–Ω—è—è –≤—ã—Ä—É—á–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ç–∞—Ä–∏—Ñ–æ–≤ Ultra –∏ Smart —Ä–∞–≤–Ω–∞.

–ï—Å–ª–∏ pvalue –±—É–¥–µ—Ç –±–æ–ª—å—à–µ, —á–µ–º —Ö–Ω–∞—á–µ–Ω–∏–µ alpha, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑–∞ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç—Å—è. –ï—Å–ª–∏ –º–µ–Ω—å—à–µ, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–µ—Ç—Å—è.

```python
sample_1=data_ultra['cost_per_month']
sample_2=data_smart['cost_per_month']
alpha=0.05 # –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

results=st.ttest_ind(sample_1, sample_2, equal_var=False)
print ('p-–∑–Ω–∞—á–µ–Ω–∏–µ=',results.pvalue)

if results.pvalue < alpha:
    display('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    display('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')

```
